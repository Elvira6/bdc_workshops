# Filtering dataframes


Filtering dataframes allow you to view the datasets in certain manner. By using filtering, you can derive some great insights about dataset.

Let's say, we want to find how many rows in the above dataframe has rainfall greater than 0.

```python
df_rain[df_rain['rainfall']>0]
```

```python
len(df_rain[df_rain['rainfall']>0])
```

Let's say we want to check rainfall grater than zero for 'RG_016' only.

```python
df_rain[(df_rain['name']=="RG_016")&(df_rain['rainfall']>0)]
```

Let's say we want to filter the rows with rainfall greater than 0.2 or it is for RG_018

```python
df_rain[(df_rain['name']=="RG_018")|(df_rain['rainfall']>0.2)]
```

Note that, in the above filter we are getting much larger number of rows, as it is OR( | ) operation, which considers either option to be true.
